```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
file_path = 'C:/Users/Nithish Kumar/Downloads/data_cleaned.csv'
aqi_data = pd.read_csv(file_path)
aqi_data['Date'] = pd.to_datetime(aqi_data['Date'])
aqi_data['MonthName'] = aqi_data['Date'].dt.strftime('%b')
monthly_aqi_avg = aqi_data.groupby('MonthName', sort=False)['AQI'].mean().reset_index()
plt.figure(figsize=(10, 6))
palette_colors = sns.color_palette("Spectral", len(monthly_aqi_avg))
chart = sns.barplot(x='MonthName', y='AQI', data=monthly_aqi_avg, palette=palette_colors)
chart.set_title('Mean AQI by Month')
chart.set_ylabel('Mean Air Quality Index (AQI)')
chart.set_xlabel('Month')
output_file = 'mean_monthly_aqi_chart.png'
plt.savefig(output_file)
plt.show()
```

```python
category_counts = aqi_data['AQI_Bucket'].value_counts().rename_axis('Category').reset_index(name='Frequency')
plt.figure(figsize=(12, 8))
category_palette = sns.color_palette("muted", len(category_counts))
barplot = sns.barplot(x='Category', y='Frequency', data=category_counts, palette=category_palette)
barplot.set_title('Frequency of AQI Bucket Categories')
barplot.set_xlabel('AQI Bucket Category')
barplot.set_ylabel('Frequency Count')
output_file_categories = 'frequency_aqi_bucket_categories'
plt.savefig(output_file_categories)
plt.show()
```

```python
pollutant_averages = aqi_data[['PM2_5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']].mean().rename_axis('Pollutant').reset_index(name='Average')
plt.figure(figsize=(14,7))
pollutant_palette = sns.color_palette("tab10", n_colors=len(pollutant_averages))
avg_plot = sns.barplot(x='Pollutant', y='Average', data=pollutant_averages, palette=pollutant_palette)
avg_plot.set_title('Average Concentration of Pollutants and AQI')
avg_plot.set_xlabel('Type of Pollutant')
avg_plot.set_ylabel('Average Concentration')
plt.savefig('average_concentration_pollutants_aqi.png')
plt.show()
```

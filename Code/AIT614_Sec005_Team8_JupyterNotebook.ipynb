```python
import pandas as pd

# Load the dataset
data = pd.read_csv('D:\AIT614\Project\data_cleaned.csv')

# Display the first few rows of the dataframe
data.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline

# Selecting a subset of columns as features for prediction
features = ['NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3']

.
pipeline = make_pipeline(
    SimpleImputer(strategy='median'),
    LinearRegression()
)

# Let's Split the data where PM2.5 is known and where it's missing
known_pm25 = data.dropna(subset=['PM2_5'])
missing_pm25 = data[data['PM2_5'].isnull()]

# Fit the model on known PM2.5 values
pipeline.fit(known_pm25[features], known_pm25['PM2_5'])

# Predict the missing PM2.5 values
predicted_pm25 = pipeline.predict(missing_pm25[features])

# Fill in the missing PM2.5 values in the original dataframe
data.loc[data['PM2_5'].isnull(), 'PM2_5'] = predicted_pm25


data.head(100)
```

```python
# Continue the same process as above to the remaining columns where data is missing
features = ['PM2_5', 'NO', 'NO2', 'NOx', 'CO', 'SO2', 'O3']

pipeline = make_pipeline(
    SimpleImputer(strategy='median'),  
    LinearRegression()
)

known_pm10 = data.dropna(subset=['PM10'])
missing_pm10 = data[data['PM10'].isnull()]

pipeline.fit(known_pm10[features], known_pm10['PM10'])

predicted_pm10 = pipeline.predict(missing_pm10[features])

data.loc[data['PM10'].isnull(), 'PM10'] = predicted_pm10

data.head()
```
